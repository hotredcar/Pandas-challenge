```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path('Resources/schools_complete.csv')
student_data_to_load = Path('Resources/students_complete.csv')
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how='left', on='school_name')
```

```python
school_data_complete_df.head()
```

```python
school_data_complete_df.describe()
```

```python
total_schools = school_data_complete_df['school_name'].nunique()
total_schools
```

```python
total_students = len(school_data_complete_df['student_name'])
total_students
```

```python
total_budget = school_data_df['budget'].sum()
total_budget
```

```python
average_math_score = school_data_complete_df['maths_score'].mean()
average_math_score
```

```python
average_reading_score = school_data_complete_df['reading_score'].mean()
average_reading_score
```

```python
passing_maths = len(school_data_complete_df[school_data_complete_df['maths_score']>=50])
percentage_passing_maths = passing_maths / total_students * 100
percentage_passing_maths
```

```python
passing_reading = len(school_data_complete_df[school_data_complete_df['reading_score']>=50])
percentage_passing_reading = passing_reading / total_students * 100
percentage_passing_reading
```

```python
passing_both = len(school_data_complete_df[(school_data_complete_df['reading_score']>=50) & (school_data_complete_df['maths_score']>=50)])
percentage_passing_both = passing_both / total_students * 100
percentage_passing_both
```

```python
area_summary = {'Total Schools':[total_schools],
               'Total Students':[total_students],
               'Total Budget':[total_budget],
               'Average Math Score':[average_math_score],
               'Average Reading Score':[average_reading_score],
               '% Paasing Maths':[percentage_passing_maths],
               '% Passing reading':[percentage_passing_reading],
               '% Overall Passing':[percentage_passing_both]}
area_summary_df = pd.DataFrame(area_summary)

area_summary_df['Total Students'] = area_summary_df['Total Students'].map('{:,}'.format)
area_summary_df['Total Budget'] = area_summary_df['Total Budget'].map('${:,.2f}'.format)

area_summary_df.head()
```

```python
school_data_complete_df['pass_maths'] = school_data_complete_df['maths_score']>=50
school_data_complete_df['pass_reading'] = school_data_complete_df['reading_score']>=50
school_data_complete_df['overall_pass'] = school_data_complete_df['pass_maths'] & school_data_complete_df['pass_reading']

percentage_pass_maths = grouped_summary_df['pass_maths'].mean()*100
percentage_pass_reading = grouped_summary_df['pass_reading'].mean()*100
percentage_overall_pass = grouped_summary_df['overall_pass'].mean()*100
percentage_pass_maths
```

```python
school_type = school_data_df.set_index("school_name")["type"]
school_type
```

```python
school_budget = school_data_df.groupby('school_name')['budget'].sum()
school_budget

student_budget = school_budget / summary_total_students
student_budget
```

```python
school_summary_df = pd.DataFrame({'School Type': school_type,
                                  'Total Students':summary_total_students,
                                  'Total School Budget':school_budget,
                                  'Per Student Budget': student_budget,
                                  'Average Maths Score': summary_average_maths,
                                  'Average Reading Score': summary_average_reading,
                                  '% Passing Maths': percentage_pass_maths,
                                  '% Passing Reading': percentage_pass_reading,
                                  '% Overall Passing':percentage_overall_pass  
                              })

school_summary_df.head()
```

```python
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map('${:,.2f}'.format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map('${:,.2f}'.format)
school_summary_df.head()
```
