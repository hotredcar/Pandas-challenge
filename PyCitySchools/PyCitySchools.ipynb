```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path('Resources/schools_complete.csv')
student_data_to_load = Path('Resources/students_complete.csv')
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how='left', on='school_name')
```

```python
school_data_complete_df.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data_complete_df['school_name'].nunique()
total_schools
```

```python
total_students = len(school_data_complete_df['student_name'])
total_students
```

```python
total_budget = school_data_df['budget'].sum()
total_budget
```

```python
average_math_score = school_data_complete_df['maths_score'].mean().round(2)
average_math_score
```

```python
average_reading_score = school_data_complete_df['reading_score'].mean().round(2)
average_reading_score
```

```python
passing_maths = len(school_data_complete_df[school_data_complete_df['maths_score']>=50])
percentage_passing_maths = round(passing_maths / total_students * 100,2)
percentage_passing_maths
```

```python
passing_reading = len(school_data_complete_df[school_data_complete_df['reading_score']>=50])
percentage_passing_reading = round(passing_reading / total_students * 100,2)
percentage_passing_reading
```

```python
passing_both = len(school_data_complete_df[(school_data_complete_df['reading_score']>=50) & (school_data_complete_df['maths_score']>=50)])
percentage_passing_both = round(passing_both / total_students * 100,2)
percentage_passing_both
```

```python
area_summary = {'Total Schools':[total_schools],
               'Total Students':[total_students],
               'Total Budget':[total_budget],
               'Average Math Score':[average_math_score],
               'Average Reading Score':[average_reading_score],
               '% Passing Maths':[percentage_passing_maths],
               '% Passing reading':[percentage_passing_reading],
               '% Overall Passing':[percentage_passing_both]}
area_summary_df = pd.DataFrame(area_summary)

area_summary_df['Total Students'] = area_summary_df['Total Students'].map('{:,}'.format)
area_summary_df['Total Budget'] = area_summary_df['Total Budget'].map('${:,.2f}'.format)

area_summary_df.head()
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data_complete_df['school_name'].unique()
```

```python
school_type = school_data_df.set_index("school_name")["type"]
```

```python
summary_average_reading = school_data_complete_df.groupby('school_name')['reading_score'].mean()
summary_average_maths = school_data_complete_df.groupby('school_name')['maths_score'].mean()
summary_total_students = school_data_complete_df.groupby('school_name')['student_name'].count()
```

```python
school_data_complete_df['pass_maths'] = school_data_complete_df['maths_score']>=50
school_data_complete_df['pass_reading'] = school_data_complete_df['reading_score']>=50
school_data_complete_df['overall_pass'] = school_data_complete_df['pass_maths'] & school_data_complete_df['pass_reading']

percentage_pass_maths = school_data_complete_df.groupby('school_name')['pass_maths'].mean()*100
percentage_pass_reading = school_data_complete_df.groupby('school_name')['pass_reading'].mean()*100
percentage_overall_pass = school_data_complete_df.groupby('school_name')['overall_pass'].mean()*100
```

```python
school_budget = school_data_df.groupby('school_name')['budget'].sum()
student_budget = school_budget / summary_total_students
```

```python
per_school_summary_df = pd.DataFrame({'School Type': school_type,
                                      'Total Students':summary_total_students,
                                      'Total School Budget':school_budget,
                                      'Per Student Budget': student_budget,
                                      'Average Maths Score': summary_average_maths,
                                      'Average Reading Score': summary_average_reading,
                                      '% Passing Maths': percentage_pass_maths,
                                      '% Passing Reading': percentage_pass_reading,
                                      '% Overall Passing':percentage_overall_pass})
```

```python
per_school_summary_df['Total School Budget'] = per_school_summary_df['Total School Budget'].map('${:,.2f}'.format)
per_school_summary_df['Per Student Budget'] = per_school_summary_df['Per Student Budget'].map('${:,.2f}'.format)
per_school_summary_df
```

## Scores by School Spending

```python
school_spending_df = pd.DataFrame({'Average Maths Score': summary_average_maths,
                                   'Average Reading Score': summary_average_reading,
                                   '% Passing Maths': percentage_pass_maths,
                                   '% Passing Reading': percentage_pass_reading,
                                   '% Overall Passing':percentage_overall_pass,
                                   'Per Student Budget': student_budget})

bins = [0,585,630,645,680]
group_range = ['<$585','$585-630','$630-645','$645-680']

school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_spending_df['Per Student Budget'], bins, labels=group_range, include_lowest=True)

spending_summary = school_spending_df.groupby('Spending Ranges (Per Student)')
spending_summary[['Average Maths Score', 'Average Reading Score','% Passing Maths','% Passing Reading','% Overall Passing']].mean().round(2)
```
